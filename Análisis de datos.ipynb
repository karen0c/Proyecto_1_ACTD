```python
import io
columnas = ["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"]
datos_iniciales = pd.read_csv('processed.cleveland.data', header=None, names=columnas, na_values="?")
datos_iniciales = datos_iniciales.dropna().reset_index(drop=True) #elimina filas con valores faltantes
print(datos_iniciales.head())
print(datos_iniciales.tail())
```

##Análisis de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###Variable Age

Datos originales:

```python
# Estadísticas descriptivas
datos_iniciales["age"].describe()
```

```python
# Histograma
datos_iniciales["age"].hist()
datos_iniciales["age"].hist().set_title('Histograma de la variable age')
plt.show()
```

Datos discretizados:

Como en el histograma se muestran muchas clases, vamos a discretizarlo de la siguiente forma:

*   1 -> edades <= 40
*   2 -> edades > 40 y <=50
*   3 -> edades > 50 y <=60
*   4 -> edades > 60

```python
# crear unos nuevos datos donde guardaremos la información con los datos discretizados
datos = datos_iniciales
```

```python
for i in range(0,297):
  if datos_iniciales.loc[i, 'age'] <= 40:
    datos.loc[i, 'age'] = 1
  elif datos_iniciales.loc[i, 'age'] > 40 and datos_iniciales.loc[i, 'age'] <= 50:
    datos.loc[i, 'age'] = 2
  elif datos_iniciales.loc[i, 'age'] > 50 and datos_iniciales.loc[i, 'age'] <= 60:
    datos.loc[i, 'age'] = 3
  else:
    datos.loc[i, 'age'] = 4
```

```python
# Estadísticas descriptivas
datos["age"].describe()
```

```python
# Histograma
datos["age"].hist()
datos["age"].hist().set_title('Histograma de la variable edad')
plt.show()

# Diagrama de caja
datos["age"].plot(kind="box", title='Diagrama de caja para la variable edad')
plt.show()

# Diagrama de dispersión
datos.plot(kind="scatter", x="age", y="num", title='Diagrama de dispersión del diagnóstico respecto a la edad')
plt.show()

# Diagrama de violín
sns.violinplot(x="age", y="num", data=datos)
plt.show()
```

### Variable trestbps

Datos originales

```python
# Estadísticas descriptivas
datos["trestbps"].describe()
```

```python
# Histograma
datos["trestbps"].hist()
datos["trestbps"].hist().set_title('Histograma para la presión arterial en reposo')
plt.show()
```

Como en el histograma se muestran muchas clases, vamos a usar las siguientes categorías:

*   1 (presión arterial normal): valores <= 120
*   2 (prehipertensión): valores > 120 y <= 139
*   3 (hipertensión grado 1): valores >= 140 y <= 159
*   4 (hipertensión grado 2): valores >= 160 y <= 179
*   5 (Crisis hipertensiva): valores >= 180


```python
for i in range(0,297):
  if datos_iniciales.loc[i, "trestbps"] <= 120:
    datos.loc[i, "trestbps"] = 1
  elif datos_iniciales.loc[i, "trestbps"] > 120 and datos_iniciales.loc[i, "trestbps"] <= 139:
    datos.loc[i, "trestbps"] = 2
  elif datos_iniciales.loc[i, "trestbps"] >= 140 and datos_iniciales.loc[i, "trestbps"] <= 159:
    datos.loc[i, "trestbps"] = 3
  elif datos_iniciales.loc[i, "trestbps"] >= 160 and datos_iniciales.loc[i, "trestbps"] <= 179:
    datos.loc[i, "trestbps"] = 4
  else:
    datos.loc[i, "trestbps"] = 5
```
